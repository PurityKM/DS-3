```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000) # x is a 1000 evenly spaced numbers from 0 to 10 (start, stop, num)
y = np.sin(x)

ax.plot(x, y)
plt.show()
```

## alternatively

```python
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('SineGraph')
plt.show()
```

```python
z = np.linspace(0, 10, 1000)
plt.plot(z, np.cos(z))
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('CosineGraph')
plt.show()
```

```python
plt.plot(z, np.sin(z), color='k', label='sin(z)')
plt.plot(z, np.cos(z), color='r', label='cos(z)', linestyle='--')
plt.xlabel('X-axis', color='r')
plt.ylabel('Y-axis', color='r')
plt.title('SINE and COSINE GRAPH', color='b')
plt.legend()
plt.show()
```

```python
y1 = z**2
y2 = z
plt.plot(z, y1, label='y1')
plt.plot(z, y2, label='y2')
plt.legend()
plt.show()
```
